## P1. Descomposición sesgo-varianza.

a) Considere el conjunto de observaciones $ D = \{(x_i,y_i)\}_{i=1}^N $, relacionadas mediante el modelo lineal
    
$$ y_i = \theta^T x_i + \varepsilon_i $$

donde $\{\varepsilon_i\}_{i=1}^N$ son observaciones i.i.d. con $\mathbb{E}[\varepsilon] = 0$ y $var(\varepsilon) = \sigma^2$, $\theta$ es un parámetro fijo y desconocido. Para un nuevo par, denotado $(x,y)$, no contenido en el conjunto de observaciones $D$, considere la predicción de $y$ mediante $\hat{y} = \hat{\theta}^Tx$, donde $\hat{\theta}$ es un estimador del parámetro $\theta$ basado en $D$. Muestre que el costo cuadrático esperado de predeccir $y$ con $\hat{y}$ (recuerde que las esperanzas se toman con respecto a la ley de $\varepsilon$) admite la siguiente descomposición sesgo-varianza:
    
$$ \mathbb{E}\left[ (\hat{y} -y)^2 \right]\, =\, var(\hat{y}) + sesgo(\hat{y})^2 + \sigma^2, $$

donde: i) el primer término es la varianza de la variable aleatoria $\hat{y}$, con $var(\hat{y}) = \mathbb{E}[(\hat{y} - \mathbb{E}[\hat{y}])^2]$ (recuerde que $\hat{\theta}$ es una variable aleatoria, pues depende de $D$), ii) el segundo término es el sesgo al cuadrado de la variable aleatoria $\hat{y}$, con $sesgo(\hat{y}) = \mathbb{E}[\hat{y}-y] = \mathbb{E}[\hat{y}] - \theta^Tx$, y iii) $\sigma^2$ es la varianza de $\varepsilon$.

#### Demostración:

Primero notemos que: el estadístico $\hat{\theta}$ es una función del conjunto $D$ de datos, es decir $\hat{\theta}(D)$, por lo tanto la estimación $\hat{y}$ es una función de $D$ y el nuevo valor $x$, es decir $\hat{y} = \hat{y}\,(D,x) = \hat{\theta}^Tx$. Dado que $(x,y)\notin D$, $y$ es independiente a todos los valores $y_i$ en $\{(x_i,y_i)\}_{i=1}^N$ (pues las únicas variables aleatorias involucradas son los $\{\varepsilon_i\}_i$, y estas son i.i.d.), por lo tanto $y$ también es independiente a toda función que dependa sólo de este conjunto $D$, es decir $y$ e $\hat{y}$ son v.a. independientes.

Como $y$ e $\hat{y}$ son independientes, entonces $\mathbb{E}(y\hat{y}) = \mathbb{E}(y)\,\mathbb{E}(\hat{y})$.

Además, notemos que $\mathbb{E}(y) = \mathbb{E}(\theta^Tx + \varepsilon) = \theta^Tx$, por lo tanto podemos decir que $y = \mathbb{E}[y] + \varepsilon$.

Con esto tenemos que:

$\mathbb{E}[(\hat{y} - y)^2] = $

$$ \mathbb{E}\left[ \hat{y}^2\, -\, 2\,\hat{y}\,y\,+\,y^2\right]\,=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}\,y]\,+\,\mathbb{E}[y^2]$$

$$=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}[y^2]\,=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}\left[\left(\mathbb{E}(y)\,+\,\varepsilon\right)^2\right]$$

Donde las dos últimas igualdades se ha ocupado lo explicado anteriormente.

$$=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}\left[\mathbb{E}(y)^2\,+\,2\,\mathbb{E}(y)\,\varepsilon\,+\,\varepsilon^2\right]$$

$$=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}[y]^2\,+\,2\,\mathbb{E}[y]\,\mathbb{E}[\varepsilon]\,+\,\mathbb{E}[\varepsilon^2]$$

Se ha ocupado que la esperanza es un valor constante, por lo tanto sale de otra esperanza iterada. Además notemos que como $\mathbb{E}[\varepsilon] = 0$, entonces $\mathbb{E}[\varepsilon^2] = var[\varepsilon] = \sigma^2$. Con esto:
 
$$=\,\mathbb{E}[\hat{y}^2]\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}[y]^2\,+\,\sigma^2$$
 
Sólo basta restar y sumar el término $\mathbb{E}[\hat{y}]^2$ para llegar a nuestro resultado.
 
$$=\,\mathbb{E}[\hat{y}^2]\,+\,\left(-\,\mathbb{E}[\hat{y}]^2\,+\,\mathbb{E}[\hat{y}]^2\right)\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}[y]^2\,+\,\sigma^2$$
 
$$=\,\left(\mathbb{E}[\hat{y}^2]\,+\,-\,\mathbb{E}[\hat{y}]^2\right)\,+\,\left(\mathbb{E}[\hat{y}]^2\,-\,2\,\mathbb{E}[\hat{y}]\,\mathbb{E}[y]\,+\,\mathbb{E}[y]^2\right)\,+\,\sigma^2$$
 
$$=\,var[\hat{y}]\,+\,\left(\mathbb{E}[\hat{y}]\,-\,\mathbb{E}[y]\right)^2\,+\,\sigma^2$$
 
$$=\,var[\hat{y}]\,+\,\left(\mathbb{E}[\hat{y}\,-\,y]\right)^2\,+\,\sigma^2$$
 
$$=\,var[\hat{y}]\,+\,\left(sesgo(\hat{y})\right)^2\,+\,\sigma^2$$

b) Para los parámetros de mínimos cuadrados $\theta_{MC} = \left(X^TX\right)^{-1}X^TY$ y de mínimos cuadrados regularizados $\theta_{MCR} = \left(X^TX + \rho I\right)^{-1}X^TY$ calcule el sesgo y la varianza de la predicción.

Primero notemos que para una nueva matriz de observaciones $\tilde{X}$, el modelo se escribe como:
 
$$ Y\, =\, \tilde{X}\cdot\theta\,+\,\mathcal{E}$$
 
donde $\cdot$ es el producto matricial y $\mathcal{E}$ es el vector de errores aleatorios, y:
 
$$\mathbb{E}(Y) = \tilde{X}\cdot\theta$$
 
#### Para mínimos cuadrados:
 
$sesgo(\hat{Y}_{MC}) = $
 
$$ \mathbb{E}[\hat{Y}_{MC}] - \tilde{X}\cdot\theta = \mathbb{E}\left[\tilde{X}\cdot\theta_{MC}\right] - \tilde{X}\cdot\theta $$
 
$$ =\,\tilde{X}\cdot\mathbb{E}\left[\theta_{MC}\right]\,-\,\tilde{X}\cdot\theta $$
 
$$ =\,X\cdot\mathbb{E}\left[\left(X^TX\right)^{-1}X^T\,Y\right]\,-\,X\cdot\theta $$
 
$$ =\,\tilde{X}\left(X^TX\right)^{-1}X^T\cdot\mathbb{E}\left[\,Y\right]\,-\,\tilde{X}\cdot\theta $$
 
$$ =\,\tilde{X}\left[\left(X^TX\right)^{-1}X^TX\right]\cdot\theta\,-\,\tilde{X}\cdot\theta $$
 
$$ = \,\tilde{X}\cdot\theta\,-\,\tilde{X}\cdot\theta\,=\,0$$


Como ya notamos, la predicción $\hat{Y}_{MC}$ es insesgada con respecto al valor esperado de $Y$, por lo tanto $\mathbb{E}(\hat{Y}_{MC}) = X\cdot\theta$. Además, de la misma forma concluímos que $\hat{\theta}$ es estimador insesgado de $\theta$.

Para calcular la varianza, notemos lo siguiente. Partiendo del modelo inicial:

$$Y\,=\,X\cdot \theta\,+\,\mathcal{E}$$

Multiplicando por el factor $\left(X^TX\right)^{-1}X^T$ a ambos lados de la igualdad:

$$\left(X^TX\right)^{-1}X^T\cdot Y\,=\,\left(X^TX\right)^{-1}X^T\cdot X\cdot \theta\,+\,\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}$$

Así, notamos que el primer lado de la igualdad es el estimador de mínimos cuadrados $\theta_{MC} = \left(X^TX\right)^{-1}X^T\,Y$, y el término que viene inmediatamente luego de la igualdad se cancela, pues $\left(X^TX\right)^{-1}X^T\cdot X\,=\,I$, donde $I$ es la identidad del tamaño igual a la cantidad de parámetros. De esta forma, la igualdad anterior queda de la forma:

$$\theta_{MC}\,=\,\theta\,+\,\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}$$

$$\Longrightarrow\,\,\theta_{MC}\,-\,\theta\,=\,\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}$$

Con esto calculemos:

$var(\hat{Y}_{MC})\,=$

$$var(\tilde{X}\cdot \theta_{MC}) = \tilde{X}\cdot var(\theta_{MC})\cdot \tilde{X}^T$$

De esta forma sólo debemos calcular la varianza del estadístico $\theta_{MC}$: 

$var(\theta_{MC}) =$

$$\mathbb{E}\left[\left(\theta_{MC}\,-\,\mathbb{E}(\theta_{MC})\right)\left(\theta_{MC}\,-\,\mathbb{E}(\theta_{MC})\right)^T\right]\,=\,\mathbb{E}\left[\left(\theta_{MC}\,-\,\theta\right)\left(\theta_{MC}\,-\,\theta\right)^T\right]$$

$$=\,\mathbb{E}\left[\left(\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}\right)\left(\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}\right)^T\right]$$

$$=\,\mathbb{E}\left[\left(X^TX\right)^{-1}X^T\cdot \mathcal{E}\,\cdot\mathcal{E}^T\,X\left(X^TX\right)^{-1}\right]$$

$$=\,\left(X^TX\right)^{-1}X^T\cdot\mathbb{E}\left[\mathcal{E}\,\mathcal{E}^T\right]\cdot\,X\left(X^TX\right)^{-1}$$

Donde el término que queda al medio, $\mathbb{E}[\mathcal{E}\mathcal{E}^T]$ es la matriz de varianzas y covarianzas del vector de errores $\mathcal{E}$. Puesto que los errores son i.i.d, esta matriz no es más que $\sigma^2\,I$. Por lo tanto:

$$=\,\sigma^2\left(X^TX\right)^{-1}X^T\cdot\,X\left(X^TX\right)^{-1} = \sigma^2\,\left(X^TX\right)^{-1}$$

Con esto; $var(\theta_{MC})\,=\,\sigma^2\left(X^TX\right)^{-1}$. Por lo tanto:

$$var(\hat{Y}_{MC})\,=\,\sigma^2\,\tilde{X}\cdot\left(X^TX\right)^{-1}\cdot\tilde{X}^T$$

#### Para mínimos cuadrados regularizados.

$sesgo(\hat{Y}_{MCR})\,=$

$$\mathbb{E}[\hat{Y}_{MCR}]\,-\,\tilde{X}\cdot\theta\,$$ 

$$=\,\mathbb{E}[\tilde{X}\cdot\theta_{MCR}]\,-\,\tilde{X}\cdot\theta$$

$$=\,\tilde{X}\cdot\mathbb{E}[\theta_{MCR}]\,-\,\tilde{X}\cdot\theta\,$$

$$=\,\tilde{X}\cdot\mathbb{E}[\left(X^TX\,+\,\rho I\right)^{-1}X^TY]\,-\,\tilde{X}\cdot\theta$$

$$=\,\tilde{X}\cdot\left(X^TX\,+\,\rho I\right)^{-1}X^T\mathbb{E}[Y]\,-\,\tilde{X}\cdot\theta\,$$

$$=\,\tilde{X}\cdot\left(X^TX\,+\,\rho I\right)^{-1}X^T\, X\cdot\theta\,-\,\tilde{X}\cdot\theta$$

$$=\,\tilde{X}\cdot\left[\left(X^TX\,+\,\rho I\right)^{-1}X^TX\,-\,I\right]\cdot\theta$$

Habiendo factorizado correctamente, podemos reescribir la matriz indentidad como $I\,=\,\left(X^TX\,+\,\rho I\right)^{-1}\cdot\left(X^TX\,+\,\rho I\right)$. Así:

$$=\,\tilde{X}\cdot\left[\left(X^TX\,+\,\rho I\right)^{-1}X^TX\,-\,\left(X^TX\,+\,\rho I\right)^{-1}\cdot\left(X^TX\,+\,\rho I\right)\right]\cdot\theta$$

$$=\,\tilde{X}\left(X^TX\,+\,\rho I\right)^{-1}\cdot\left[X^TX\,-\,\left(X^TX\,+\,\rho I\right)\right]\cdot\theta$$

$$=\,-\rho\tilde{X}\left(X^TX\,+\,\rho I\right)^{-1}\cdot\theta$$

$var(\hat{Y}_{MCR})\,=$

$$var(\tilde{X}\cdot\theta_{MCR})\,=\,\tilde{X}\cdot var(\theta_{MCR})\cdot\tilde{X}^T$$

$$=\,\tilde{X}\cdot var\left[\left(X^TX\,+\,\rho I\right)^{-1}X^T\,Y\right]\cdot \tilde{X}^T$$

$$=\,\tilde{X}\left(X^TX\,+\,\rho I\right)^{-1}X^T\cdot var(Y)\cdot X \left(X^TX\,+\,\rho I\right)^{-1}\tilde{X}^T$$

Sólo nos falta ver la varianza de $Y$, que es la misma que el vector de errores $\mathcal{E}$ ya que la varianza no cambia por traslaciones constantes, es decir: $var(Y)\,=\,var(X\theta\,+\,\mathcal{E})\,=\,var(\mathcal{E})\,=\,\sigma^2 I$.

$$=\,\sigma^2\tilde{X}\left(X^TX\,+\,\rho I\right)^{-1}X^TX \left(X^TX\,+\,\rho I\right)^{-1}\tilde{X}^T$$

Como base de datos del experimento se utilizará el archivo **Housing.csv**, este corresponde a un data set que posee precios de casas de algunas localidades de USA. El objetivo de esta pregunta es aprender una función lineal que relaciones $X$ e $Y$, así un agente inmobiliario que no tiene información sobre el precio de las casas en una nueva localidad pueda fijarle un precio a estas en base al ingreso promedio del área.
Para una correcta comparación de los estimadores la base de datos viene dividida en dos conjuntos, entrenamiento (*in-sample*) y validación (*out-of-sample*).

Además de cargar los datos, incluiremos una nueva columna de datos a la base, la columnta *inter*, la que consta de sólo valores 1.0, esta nos ayudará a agregar un intercepto al modelo. Es decir, asumiremos que la regresión consta de una función *afín* y no, rigurosamente, *lineal*.

b) Implemente el estimador de mínimos cuadrados regularizado usando regularización *ridge* y obtenga el vector de parámetros $\theta$ para diferentes valores de $\rho\,\in\,[0,10]$ incluyendo los límites. Para esto  implemente la función $\texttt{reg_lineal(X,Y,rho)}$. 

```python
def reg_lineal(X,Y,rho):
    #X e Y deben ser vectores array de numpy
    #rho debe ser un float
    x1 = np.dot(X.T,X) #X.T @ X
    if np.shape(x1) == ():
        x2 = (x1 + rho)**(-1)
    else:
        n = np.shape(x1)[0]
        x2 = np.linalg.inv(x1 + rho*np.eye(n))
    
    x3 = np.dot(X.T,Y)
    theta = np.dot(x2,x3)
    return theta
```

```python
# separar conjunto de entrenamiento (1) y conjunto de validación (0)

train = data[data[cols[3]]==1]
test = data[data[cols[3]]==0]

# Eligiremos 500 valores entre 0 y 10 con el comando linspace
rhos = np.linspace(0,10,5000)
thetas =[]
intercep = []
X = np.array([train[cols[1]], train[cols[4]]]).T
Y = train[cols[2]]

for r in rhos:
    A = reg_lineal(X,Y,r)
    thetas.append(A[0])
    intercep.append(A[1])
```

c) Grafique el valor de los parámetros estimados para los valores de $\rho$.

```python
plt.figure(figsize = (15,7))
plt.plot(rhos,thetas)
plt.xlabel('Parámetro de regularización.')
plt.ylabel('Parámetro estimado del modelo.')
plt.title('Parámetro theta 0 en función de la regularización.')
plt.show()
```

Notamos que el parámetro $\theta_0$, correspondiente a la pendiente de la recta, disminuye de forma inversa con ell aumento del parámetro de regularización. Esto quiere decir que la recta va rotando en sentido horario, cada vez en un menor ángulo.

```python
plt.figure(figsize = (15,7))
plt.plot(rhos,intercep)
plt.xlabel('Parámetro de regularización.')
plt.ylabel('Parámetro estimado del modelo.')
plt.title('Parámetro theta 1 en función de la regularización.')
plt.show()
```

Como era de esperarse, dad la ratoción de las rectas descritas en el gráfico anterior, el parámetro de intersección aumenta para cada recta. Debemos considerar que el aumento del parámetro de regularización $\rho$, implica añadir penalización a la norma de $\theta \,=\,(\theta_0,\theta_1)$, por lo tanto el aumento de $\theta_1$ debe ser tal que $||\theta||_2$ siga en disminución, como se verá en el siguiente gráfico.

```python
norm = []
for i in range(len(thetas)):
    n = np.sqrt(thetas[i]**2 + intercep[i]**2)
    norm.append(n)

plt.figure(figsize = (15,7))
plt.plot(rhos,norm)
plt.xlabel('Parámetro de regularización.')
plt.ylabel('Norma de theta.')
plt.title('Norma de theta en función de rho.')
plt.show()
```

d) Grafique el error cuadrático medio y la varianza  de la predicción  para los valores  de $\rho$ tanto en el conjunto de entrenamiento  como en el de validación.

```python
# Para conjunto de entrenamiento:
# varianza:
varianza_train = []
for i in range(len(thetas)):
    y_est = thetas[i]*(X.T[0]) + intercep[i]
    varianza_train.append(np.var(y_est))

#ECM:
ECM_train = []
for i in range(len(thetas)):
    error = thetas[i]*(X.T[0]) + intercep[i] - Y
    ec = [a**2 for a in error]
    ECM_train.append(np.mean(ec))

plt.figure(figsize = (12,5))
plt.subplot(121)
plt.xlabel('Parámetro de Regularización')
plt.title('Varianza entrenamiento.')
plt.plot(rhos,varianza_train, '.', color = 'r', label = 'Varianza.')
plt.legend()
plt.subplot(122)
plt.xlabel('Parámetro de Regularización')
plt.title('Error Cuadrático Medio entrenamiento.')
plt.plot(rhos,ECM_train, '.', color = 'b', label = 'E.C.M.')
plt.legend()
plt.show()
```

```python
# Para conjunto de validación:
X_test = test[cols[1]]
Y_test = test[cols[2]]

# varianza:
varianza_test = []
for i in range(len(thetas)):
    y_est = thetas[i]*X_test + intercep[i]
    varianza_test.append(np.var(y_est))

#ECM:
ECM_test = []
for i in range(len(thetas)):
    error = thetas[i]*X_test + intercep[i] - Y_test
    ec = [a**2 for a in error]
    ECM_test.append(np.mean(ec))

plt.figure(figsize = (12,5))
plt.subplot(121)
plt.xlabel('Parámetro de Regularización')
plt.title('Varianza validación.')
plt.plot(rhos,varianza_test, '*', color = 'r', label = 'Varianza.')
plt.legend()
plt.subplot(122)
plt.xlabel('Parámetro de Regularización')
plt.title('Error Cuadrático Medio validación.')
plt.plot(rhos,ECM_test, '*', color = 'b', label = 'E.C.M.')
plt.legend()
plt.show()
```

En ambos casos notamos una disminución de la varianza de las estimaciones, acompañada de un aumento del error cuadrático medio, lo que implica (por lo ya demostrado) de un aumento del sesgo en las estimaciones. Sin embargo, en el conjunto de validación se puede notar descenso del *EMC* entre los parámetros 0 y 2 de $\rho$, lo que significa que existe un valor que minimiza la suma entre la varianza y el cuadrado del sesgo para las estimaciones respecto al conjunto *testing*. Será ese valor que consideraremos al elegir una de estas regresiones como la definitiva en el uso para solucionar el problema de estimación.

```python
m = np.min(ECM_test)
r = [rhos[i] for i in range(len(rhos)) if ECM_test[i] == m][0]

print('El ECM mínimo en el conjunto Test es de ',m)
print(' ')
print('Y se encuentra en el valor de regularización ',r)
```

```python
#Crearemos un nuevo conjunto de valores de rho que sea más pequeño y graficable

[a_opt,b_opt] = reg_lineal(X,Y,0.8502)

rho2 = np.linspace(0,10,5)
thetas2 = []
inter2 = []
for r in rho2:
    A = reg_lineal(X,Y,r)
    thetas2.append(A[0])
    inter2.append(A[1])
    

plt.figure(figsize = (15,15))
plt.plot(avg,price, '*', color = 'k')
for i in range(len(thetas2)):
    y = thetas2[i]*avg + inter2[i]
    plt.plot(avg,y, label = 'rho = '+str(rho2[i]))
plt.plot(avg,a_opt*avg + b_opt , 'o-',label = 'rho óptimo')
plt.legend()
plt.title('Comparación de modelos regularizados.')
plt.xlabel('Ingreso promedio.')
plt.ylabel('Precio')
plt.show()
```
